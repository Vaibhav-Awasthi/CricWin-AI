```python
total_score_df = delivery.groupby(['match_id', 'inning']).sum()['total_runs'].reset_index()
```

```python

total_score_df = total_score_df[total_score_df['inning'] == 1]
```

```python
match_df = match.merge(total_score_df[['match_id', 'total_runs']], left_on='id', right_on='match_id')
```

```python
match_df.columns
```

```python
match_df['team1'].unique()
```

```python
teams = ['Sunrisers Hyderabad', 'Mumbai Indians', 'Royal Challengers Bangalore',
       'Kolkata Knight Riders', 'Kings XI Punjab',
       'Chennai Super Kings', 'Rajasthan Royals','Delhi Capitals']
teams
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils', 'Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
```

```python
match_df.head()
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df.head()
```

```python
match_df.columns
```

```python
match_df = match_df[match_df['dl_applied'] == 0]
match_df = match_df[['match_id', 'city', 'winner', 'total_runs']]
match_df.head()
```

```python
match_df
```

```python
delivery.head()
```

```python
delivery_df =  match_df.merge(delivery, on='match_id')
delivery_df.head()
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
delivery_df.head()
```

```python
delivery_df.columns
```

```python
# Apply cumsum() after grouping by 'match_id'
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df.head()
```

```python
delivery_df.tail()
```

```python
delivery_df['runs_left']=delivery_df['total_runs_x']-delivery_df['current_score']
```

```python
delivery_df.head()
```

```python
delivery_df['balls_left']= 126 - (delivery_df['over']*6 + delivery_df['ball'])
delivery_df.head()
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df.tail()
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x=='0' else "1")
delivery_df.tail()
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
```

```python
wicket = delivery_df.groupby('match_id')['player_dismissed'].cumsum().values
```

```python
wicket
```

```python
delivery_df['wicket'] = 10-wicket
```

```python
delivery_df.tail()
```

```python
# current run rate
delivery_df['crr']=  (delivery_df['current_score']*6)/(120-delivery_df['balls_left'])
delivery_df.head()
```

```python
# required run rate
delivery_df['rrr'] = (delivery_df.get('runs_left', 0) * 6) / delivery_df.get('balls_left', 1)

delivery_df.tail()
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
delivery_df['result'] = delivery_df.apply(result, axis=1)
delivery_df.head()
```

```python
final_df = delivery_df[['batting_team', 'bowling_team', 'city', 'runs_left', 'balls_left', 'wicket', 'total_runs_x', 'crr', 'rrr', 'result']]
final_df
```

```python
final_df.describe()
```

```python
final_df.shape
```

```python
final_df = final_df.sample(final_df.shape[0])
final_df.sample()
```

```python
final_df.head()
```

```python
final_df.dropna(inplace=True)
final_df.shape
```

```python
final_df = final_df[final_df['balls_left'] != 0]
final_df.describe()
```

```python
x = final_df.iloc[:, :-1]
y = final_df.iloc[:,-1]
y.head()
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
x_train
```

```python
x_test
```

```python
x_train.info()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

# Define column transformer for one-hot encoding
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')

# Create pipeline with logistic regression
pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', LogisticRegression(solver='liblinear'))
])

# Fit the pipeline (ensure x_train and y_train are defined)
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', RandomForestClassifier())
])
pipe.fit(x_train, y_train)
```
